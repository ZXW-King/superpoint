```python
df1 = pd.read_csv('../105output.txt',names=["图片", "数量"])
df2 = pd.read_csv('../106output.txt',names=["图片", "数量"])
df3 = pd.read_csv('../107output.txt',names=["图片", "数量"])
df4 = pd.read_csv('../coco_test_magic.txt',names=["图片", "数量"])
```
